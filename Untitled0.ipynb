```python
import numpy as np
from scipy.stats import norm

# Datos proporcionados
desviacion_estandar_poblacional = 1.2  # kg
margen_error = 0.5  # kg
nivel_confianza = 0.99

# Nivel de significancia (alpha)
alpha = 1 - nivel_confianza

# Valor crítico Z para un intervalo de confianza del 99%
valor_critico_z = norm.ppf(1 - alpha/2)

# Calcular el tamaño de muestra necesario
# Fórmula: n = (Z^2 * sigma^2) / ME^2
n_necesario = (valor_critico_z**2 * desviacion_estandar_poblacional**2) / (margen_error**2)

# Redondear hacia arriba para asegurar el margen de error
n_necesario = np.ceil(n_necesario)

# Imprimir resultados
print(f"Desviación estándar poblacional (sigma): {desviacion_estandar_poblacional} kg")
print(f"Margen de error deseado: {margen_error} kg")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Valor crítico Z: {valor_critico_z:.4f}")
print(f"Tamaño de muestra necesario: {int(n_necesario)}")
```

## Prueba de Hipótesis para la Varianza del Grosor de Piezas Metálicas

El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de al menos $\sigma^2 = 0.5$ mm². Una muestra de $n = 12$ piezas dio una varianza $s^2 = 0.3$. ¿Se puede concluir al 5 % que la varianza poblacional es menor que la garantizada?

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**
2.  **Establecer el nivel de significancia α.**
3.  **Identificar el estadístico de prueba y su distribución.**
4.  **Determinar el valor crítico y la región de rechazo.**
5.  **Calcular el estadístico de prueba y el valor p.**
6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**
7.  **Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.**

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La varianza poblacional es igual a 0.5 mm² (σ² = 0.5)
# H1: La varianza poblacional es menor a 0.5 mm² (σ² < 0.5)
# Esta es una prueba de cola izquierda.
sigma2_0 = 0.5  # Varianza poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): σ² = {sigma2_0} mm²")
print(f"Hipótesis Alternativa (H1): σ² < {sigma2_0} mm² (Prueba de cola izquierda)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# Para probar la varianza poblacional con una muestra, usamos el estadístico chi-cuadrado (χ²)
# y la distribución chi-cuadrado.
print("\nEstadístico de prueba: χ²")
print("Distribución: Chi-cuadrado")

# Datos de la muestra
n = 12  # Tamaño de la muestra
s2 = 0.3  # Varianza muestral

# Grados de libertad
grados_libertad = n - 1
print(f"Grados de libertad: {grados_libertad}")

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de cola izquierda con α = 0.05 y {grados_libertad} grados de libertad,
# el valor crítico χ² es el valor que deja el 5% del área en la cola izquierda
# de la distribución chi-cuadrado.
valor_critico_chi2 = chi2.ppf(alpha, grados_libertad)
print(f"\nValor crítico χ² para α = {alpha} (cola izquierda) y {grados_libertad} GL: {valor_critico_chi2:.4f}")
print(f"Región de rechazo: χ²_calculado < {valor_critico_chi2:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²₀
estadistico_chi2 = ((n - 1) * s2) / sigma2_0
print(f"\nEstadístico χ² calculado: {estadistico_chi2:.4f}")

# Calcular el valor p para una prueba de cola izquierda
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo
# o más extremo que el observado, asumiendo que H0 es verdadera.
valor_p = chi2.cdf(estadistico_chi2, grados_libertad)
print(f"Valor p (para prueba de cola izquierda): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva chi-cuadrado
# Los valores de x van desde 0 hasta un valor que cubra la cola derecha relevante
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (GL={grados_libertad})')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(0, valor_critico_chi2, 100)
y_rechazo = chi2.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico χ²
plt.axvline(valor_critico_chi2, color='red', linestyle='--', label=f'Valor Crítico χ² = {valor_critico_chi2:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución χ² con Región de Rechazo y Estadístico χ² (GL={grados_libertad})')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_chi2 < valor_critico_chi2:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) cae en la región de rechazo (χ² < {valor_critico_chi2:.4f}).")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la varianza poblacional del grosor de las piezas metálicas es significativamente menor a {sigma2_0} mm²."
else:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) no cae en la región de rechazo (χ² < {valor_critico_chi2:.4f}).")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la varianza poblacional del grosor de las piezas metálicas es significativamente menor a {sigma2_0} mm²."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una varianza muestral tan o más extrema que {s2} mm² (es decir, menor), si la hipótesis nula (la varianza poblacional real es {sigma2_0} mm²) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una varianza muestral tan baja si la varianza poblacional real fuera {sigma2_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la varianza muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la varianza poblacional real sea {sigma2_0} mm².")
```

## Prueba de Hipótesis para la Varianza del Llenado de Botellas

Una máquina embotelladora debe mantener la varianza del llenado en $\sigma^2 = 1.2$ ml². Se toma una muestra de $n = 20$ botellas y se encuentra $s^2 = 2.0$. ¿Existe evidencia al 1 % de que la varianza es mayor?

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**
2.  **Establecer el nivel de significancia α.**
3.  **Identificar el estadístico de prueba y su distribución.**
4.  **Determinar el valor crítico y la región de rechazo.**
5.  **Calcular el estadístico de prueba y el valor p.**
6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**
7.  **Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.**

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La varianza poblacional es igual a 1.2 (σ² = 1.2)
# H1: La varianza poblacional es mayor a 1.2 (σ² > 1.2)
# Esta es una prueba de cola derecha.
sigma2_0 = 1.2  # Varianza poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): σ² = {sigma2_0} ml²")
print(f"Hipótesis Alternativa (H1): σ² > {sigma2_0} ml² (Prueba de cola derecha)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.01
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# Para probar la varianza poblacional con una muestra, usamos el estadístico chi-cuadrado (χ²)
# y la distribución chi-cuadrado.
print("\nEstadístico de prueba: χ²")
print("Distribución: Chi-cuadrado")

# Datos de la muestra
n = 20  # Tamaño de la muestra
s2 = 2.0  # Varianza muestral

# Grados de libertad
grados_libertad = n - 1
print(f"Grados de libertad: {grados_libertad}")

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de cola derecha con α = 0.01 y {grados_libertad} grados de libertad,
# el valor crítico χ² es el valor que deja el 1% del área en la cola derecha
# de la distribución chi-cuadrado.
valor_critico_chi2 = chi2.ppf(1 - alpha, grados_libertad)
print(f"\nValor crítico χ² para α = {alpha} (cola derecha) y {grados_libertad} GL: {valor_critico_chi2:.4f}")
print(f"Región de rechazo: χ²_calculado > {valor_critico_chi2:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²₀
estadistico_chi2 = ((n - 1) * s2) / sigma2_0
print(f"\nEstadístico χ² calculado: {estadistico_chi2:.4f}")

# Calcular el valor p para una prueba de cola derecha
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo
# o más extremo que el observado, asumiendo que H0 es verdadera.
valor_p = 1 - chi2.cdf(estadistico_chi2, grados_libertad)
print(f"Valor p (para prueba de cola derecha): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva chi-cuadrado
# Los valores de x van desde 0 hasta un valor que cubra la cola derecha relevante
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (GL={grados_libertad})')

# Rellenar la región de rechazo (cola derecha)
x_rechazo = np.linspace(valor_critico_chi2, chi2.ppf(0.999, grados_libertad), 100)
y_rechazo = chi2.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico χ²
plt.axvline(valor_critico_chi2, color='red', linestyle='--', label=f'Valor Crítico χ² = {valor_critico_chi2:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución χ² con Región de Rechazo y Estadístico χ² (GL={grados_libertad})')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_chi2 > valor_critico_chi2:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) cae en la región de rechazo (χ² > {valor_critico_chi2:.4f}).")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la varianza poblacional del llenado de botellas es significativamente mayor a {sigma2_0} ml²."
else:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) no cae en la región de rechazo (χ² > {valor_critico_chi2:.4f}).")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la varianza poblacional del llenado de botellas es significativamente mayor a {sigma2_0} ml²."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una varianza muestral tan o más extrema que {s2} ml² (es decir, mayor), si la hipótesis nula (la varianza poblacional real es {sigma2_0} ml²) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una varianza muestral tan alta si la varianza poblacional real fuera {sigma2_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la varianza muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la varianza poblacional real sea {sigma2_0} ml².")
```

## Prueba de Hipótesis para la Varianza del Tiempo de Duración de las Llamadas

Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de $\sigma^2 = 2.5$ minutos². Se toma una muestra de $n = 16$ llamadas y se obtiene una varianza muestral de $s^2 = 4$. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**
2.  **Establecer el nivel de significancia α.**
3.  **Identificar el estadístico de prueba y su distribución.**
4.  **Determinar los valores críticos y la región de rechazo.**
5.  **Calcular el estadístico de prueba y el valor p.**
6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, los valores críticos y el estadístico calculado.**
7.  **Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.**

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La varianza poblacional es igual a 2.5 (σ² = 2.5)
# H1: La varianza poblacional es diferente a 2.5 (σ² ≠ 2.5)
# Esta es una prueba de dos colas.
sigma2_0 = 2.5  # Varianza poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): σ² = {sigma2_0} minutos²")
print(f"Hipótesis Alternativa (H1): σ² ≠ {sigma2_0} minutos² (Prueba de dos colas)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# Para probar la varianza poblacional con una muestra, usamos el estadístico chi-cuadrado (χ²)
# y la distribución chi-cuadrado.
print("\nEstadístico de prueba: χ²")
print("Distribución: Chi-cuadrado")

# Datos de la muestra
n = 16  # Tamaño de la muestra
s2 = 4  # Varianza muestral

# Grados de libertad
grados_libertad = n - 1
print(f"Grados de libertad: {grados_libertad}")

# --- Paso 4: Determinar los valores críticos y la región de rechazo ---
# Para una prueba de dos colas con α = 0.05 y {grados_libertad} grados de libertad,
# los valores críticos χ² son los valores que dejan el α/2 = 0.025 del área en cada cola
# de la distribución chi-cuadrado.
valor_critico_chi2_inferior = chi2.ppf(alpha/2, grados_libertad)
valor_critico_chi2_superior = chi2.ppf(1 - alpha/2, grados_libertad)
print(f"\nValores críticos χ² para α = {alpha} (dos colas) y {grados_libertad} GL: {valor_critico_chi2_inferior:.4f} y {valor_critico_chi2_superior:.4f}")
print(f"Región de rechazo: χ²_calculado < {valor_critico_chi2_inferior:.4f} o χ²_calculado > {valor_critico_chi2_superior:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²₀
estadistico_chi2 = ((n - 1) * s2) / sigma2_0
print(f"\nEstadístico χ² calculado: {estadistico_chi2:.4f}")

# Calcular el valor p para una prueba de dos colas
# El valor p es el doble de la probabilidad en la cola correspondiente al estadístico χ² calculado.
# Si el estadístico calculado está a la izquierda del centro, calculamos la probabilidad de la cola izquierda y la multiplicamos por 2.
# Si el estadístico calculado está a la derecha del centro, calculamos la probabilidad de la cola derecha y la multiplicamos por 2.
# El centro de la distribución chi-cuadrado es aproximadamente igual a los grados de libertad para GL > 2.
# En este caso, 15 > 2, por lo que el centro es aproximadamente 15.
# Como el estadístico calculado (24.0000) es mayor que el centro (15), calculamos la probabilidad de la cola derecha.
if estadistico_chi2 > grados_libertad: # Aproximación para determinar la cola para el valor p en prueba de dos colas
     valor_p = 2 * (1 - chi2.cdf(estadistico_chi2, grados_libertad))
else:
     valor_p = 2 * chi2.cdf(estadistico_chi2, grados_libertad)

print(f"Valor p (para prueba de dos colas): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valores críticos y estadístico calculado ---
# Generar puntos para la curva chi-cuadrado
# Los valores de x van desde 0 hasta un valor que cubra la cola derecha relevante
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (GL={grados_libertad})')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(0, valor_critico_chi2_inferior, 100)
y_rechazo_inferior = chi2.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo Inferior (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_chi2_superior, chi2.ppf(0.999, grados_libertad), 100)
y_rechazo_superior = chi2.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.3, label=f'Región de Rechazo Superior (α/2 = {alpha/2})')


# Marcar los valores críticos χ²
plt.axvline(valor_critico_chi2_inferior, color='red', linestyle='--', label=f'Valor Crítico χ² Inferior = {valor_critico_chi2_inferior:.4f}')
plt.axvline(valor_critico_chi2_superior, color='red', linestyle='--', label=f'Valor Crítico χ² Superior = {valor_critico_chi2_superior:.4f}')


# Marcar el estadístico χ² calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución χ² con Regiones de Rechazo y Estadístico χ² (GL={grados_libertad})')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_chi2 < valor_critico_chi2_inferior or estadistico_chi2 > valor_critico_chi2_superior:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) cae en la región de rechazo.")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente a {sigma2_0} minutos²."
else:
    print(f"El estadístico χ² calculado ({estadistico_chi2:.4f}) no cae en la región de rechazo.")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente a {sigma2_0} minutos²."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una varianza muestral tan o más extrema que {s2} minutos² (en cualquier dirección desde {sigma2_0}), si la hipótesis nula (la varianza poblacional real es {sigma2_0} minutos²) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una varianza muestral tan diferente de {sigma2_0} si la varianza poblacional real fuera {sigma2_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la varianza muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la varianza poblacional real sea {sigma2_0} minutos².")
```

## Prueba de Hipótesis para la Duración Promedio de los Focos

Un fabricante asegura que sus focos duran 10,000 horas. En 20 focos probados, la media fue 9,800 horas con s = 300 horas. Con un nivel de significancia de α = 0,05.

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**
2.  **Establecer el nivel de significancia α.**
3.  **Identificar el estadístico de prueba y su distribución.**
4.  **Determinar el valor crítico y la región de rechazo.**
5.  **Calcular el estadístico de prueba y el valor p.**
6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**
7.  **Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.**

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La duración promedio de los focos es igual a 10000 horas (μ = 10000)
# H1: La duración promedio de los focos es diferente a 10000 horas (μ ≠ 10000)
# Esta es una prueba de dos colas, ya que no se especifica si se sospecha que es mayor o menor.
mu_0 = 10000  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} horas")
print(f"Hipótesis Alternativa (H1): μ ≠ {mu_0} horas (Prueba de dos colas)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es desconocida y n < 30, por lo tanto, usamos el estadístico t
# y la distribución t de Student.
print("\nEstadístico de prueba: t")
print("Distribución: t de Student")

# Datos de la muestra
n = 20  # Tamaño de la muestra
media_muestral = 9800  # horas
desviacion_estandar_muestral = 300  # horas

# Grados de libertad
grados_libertad = n - 1
print(f"Grados de libertad: {grados_libertad}")

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de dos colas con α = 0.05 y {grados_libertad} grados de libertad,
# los valores críticos t son los valores que dejan el α/2 = 0.025 del área en cada cola
# de la distribución t de Student.
valor_critico_t_inferior = t.ppf(alpha/2, grados_libertad)
valor_critico_t_superior = t.ppf(1 - alpha/2, grados_libertad)
print(f"\nValores críticos t para α = {alpha} (dos colas) y {grados_libertad} GL: {valor_critico_t_inferior:.4f} y {valor_critico_t_superior:.4f}")
print(f"Región de rechazo: t_calculado < {valor_critico_t_inferior:.4f} o t_calculado > {valor_critico_t_superior:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico t: t = (x̄ - μ0) / (s / √n)
error_estandar = desviacion_estandar_muestral / np.sqrt(n)
estadistico_t = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} horas")
print(f"Estadístico t calculado: {estadistico_t:.4f}")

# Calcular el valor p para una prueba de dos colas
# El valor p es el doble de la probabilidad en la cola correspondiente al estadístico t calculado.
# Como es de dos colas, calculamos la probabilidad en una cola y la multiplicamos por 2.
# Usamos abs(estadistico_t) para encontrar la probabilidad en la cola más extrema.
valor_p = 2 * t.cdf(-abs(estadistico_t), grados_libertad)
print(f"Valor p (para prueba de dos colas): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva t de Student
x = np.linspace(t.ppf(0.0001, grados_libertad), t.ppf(0.9999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t de Student (GL={grados_libertad})')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(t.ppf(0.0001, grados_libertad), valor_critico_t_inferior, 100)
y_rechazo_inferior = t.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo Inferior (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_t_superior, t.ppf(0.9999, grados_libertad), 100)
y_rechazo_superior = t.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.3, label=f'Región de Rechazo Superior (α/2 = {alpha/2})')


# Marcar los valores críticos t
plt.axvline(valor_critico_t_inferior, color='red', linestyle='--', label=f'Valor Crítico t Inferior = {valor_critico_t_inferior:.4f}')
plt.axvline(valor_critico_t_superior, color='red', linestyle='--', label=f'Valor Crítico t Superior = {valor_critico_t_superior:.4f}')


# Marcar el estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado = {estadistico_t:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución t de Student con Regiones de Rechazo y Estadístico t (GL={grados_libertad})')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_t < valor_critico_t_inferior or estadistico_t > valor_critico_t_superior:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) cae en la región de rechazo.")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la duración promedio real de los focos es significativamente diferente a {mu_0} horas."
else:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) no cae en la región de rechazo.")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la duración promedio real de los focos es significativamente diferente a {mu_0} horas."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral tan o más extrema que {media_muestral} horas (en cualquier dirección desde {mu_0}), si la hipótesis nula (la duración promedio real es {mu_0} horas) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan diferente de {mu_0} si la duración promedio real fuera {mu_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la duración promedio real sea {mu_0} horas.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: El consumo promedio es igual a 15 km/L (μ = 15)
# H1: El consumo promedio es diferente a 15 km/L (μ ≠ 15) - Usaremos una prueba de dos colas
# ya que no se especifica si se sospecha que es mayor o menor. Si se quisiera probar si es menor,
# sería de cola izquierda.
mu_0 = 15  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} km/L")
print(f"Hipótesis Alternativa (H1): μ ≠ {mu_0} km/L (Prueba de dos colas)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es desconocida y n < 30, por lo tanto, usamos el estadístico t
# y la distribución t de Student.
print("\nEstadístico de prueba: t")
print("Distribución: t de Student")

# Datos de la muestra
n = 10  # Tamaño de la muestra
media_muestral = 14  # km/L
desviacion_estandar_muestral = 1.2  # km/L

# Grados de libertad
grados_libertad = n - 1
print(f"Grados de libertad: {grados_libertad}")

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de dos colas con α = 0.05 y {grados_libertad} grados de libertad,
# los valores críticos t son los valores que dejan el α/2 = 0.025 del área en cada cola
# de la distribución t de Student.
valor_critico_t_inferior = t.ppf(alpha/2, grados_libertad)
valor_critico_t_superior = t.ppf(1 - alpha/2, grados_libertad)
print(f"\nValores críticos t para α = {alpha} (dos colas) y {grados_libertad} GL: {valor_critico_t_inferior:.4f} y {valor_critico_t_superior:.4f}")
print(f"Región de rechazo: t_calculado < {valor_critico_t_inferior:.4f} o t_calculado > {valor_critico_t_superior:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico t: t = (x̄ - μ0) / (s / √n)
error_estandar = desviacion_estandar_muestral / np.sqrt(n)
estadistico_t = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} km/L")
print(f"Estadístico t calculado: {estadistico_t:.4f}")

# Calcular el valor p para una prueba de dos colas
# El valor p es el doble de la probabilidad en la cola correspondiente al estadístico t calculado.
# Como es de dos colas, calculamos la probabilidad en una cola y la multiplicamos por 2.
# Usamos abs(estadistico_t) para encontrar la probabilidad en la cola más extrema.
valor_p = 2 * t.cdf(-abs(estadistico_t), grados_libertad)
print(f"Valor p (para prueba de dos colas): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva t de Student
x = np.linspace(t.ppf(0.0001, grados_libertad), t.ppf(0.9999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t de Student (GL={grados_libertad})')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(t.ppf(0.0001, grados_libertad), valor_critico_t_inferior, 100)
y_rechazo_inferior = t.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo Inferior (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_t_superior, t.ppf(0.9999, grados_libertad), 100)
y_rechazo_superior = t.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.3, label=f'Región de Rechazo Superior (α/2 = {alpha/2})')


# Marcar los valores críticos t
plt.axvline(valor_critico_t_inferior, color='red', linestyle='--', label=f'Valor Crítico t Inferior = {valor_critico_t_inferior:.4f}')
plt.axvline(valor_critico_t_superior, color='red', linestyle='--', label=f'Valor Crítico t Superior = {valor_critico_t_superior:.4f}')


# Marcar el estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado = {estadistico_t:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución t de Student con Regiones de Rechazo y Estadístico t (GL={grados_libertad})')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_t < valor_critico_t_inferior or estadistico_t > valor_critico_t_superior:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) cae en la región de rechazo.")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que el consumo promedio real de combustible es significativamente diferente a {mu_0} km/L."
else:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) no cae en la región de rechazo.")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que el consumo promedio real de combustible es significativamente diferente a {mu_0} km/L."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral tan o más extrema que {media_muestral} km/L (en cualquier dirección desde {mu_0}), si la hipótesis nula (el consumo promedio real es {mu_0} km/L) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan diferente de {mu_0} si el consumo promedio real fuera {mu_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que el consumo promedio real sea {mu_0} km/L.")
```

## Prueba de Hipótesis para el Tiempo Promedio de Entrega a Domicilio

Un restaurante afirma que su tiempo promedio de entrega a domicilio es 25 minutos. Un cliente sospecha que es mayor. Se registran 15 entregas, obteniendo una media de 28 minutos y desviación estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**
2.  **Establecer el nivel de significancia α.**
3.  **Identificar el estadístico de prueba y su distribución.**
4.  **Determinar el valor crítico y la región de rechazo.**
5.  **Calcular el estadístico de prueba y el valor p.**
6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**
7.  **Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.**

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: El tiempo promedio de entrega es igual a 25 minutos (μ = 25)
# H1: El tiempo promedio de entrega es mayor a 25 minutos (μ > 25)
# Esta es una prueba de cola derecha.
mu_0 = 25  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} minutos")
print(f"Hipótesis Alternativa (H1): μ > {mu_0} minutos (Prueba de cola derecha)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es desconocida y n < 30, por lo tanto, usamos el estadístico t
# y la distribución t de Student.
print("\nEstadístico de prueba: t")
print("Distribución: t de Student")

# Datos de la muestra
n = 15  # Tamaño de la muestra
media_muestral = 28  # minutos
desviacion_estandar_muestral = 4  # minutos

# Grados de libertad
grados_libertad = n - 1
print(f"Grados de libertad: {grados_libertad}")

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de cola derecha con α = 0.05 y {grados_libertad} grados de libertad,
# el valor crítico t es el valor que deja el 5% del área en la cola derecha de la distribución t de Student.
valor_critico_t = t.ppf(1 - alpha, grados_libertad)
print(f"\nValor crítico t para α = {alpha} (cola derecha) y {grados_libertad} GL: {valor_critico_t:.4f}")
print(f"Región de rechazo: t_calculado > {valor_critico_t:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico t: t = (x̄ - μ0) / (s / √n)
error_estandar = desviacion_estandar_muestral / np.sqrt(n)
estadistico_t = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} minutos")
print(f"Estadístico t calculado: {estadistico_t:.4f}")

# Calcular el valor p para una prueba de cola derecha
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo
# o más extremo que el observado, asumiendo que H0 es verdadera.
valor_p = 1 - t.cdf(estadistico_t, grados_libertad)
print(f"Valor p (para prueba de cola derecha): {valor_p:.4f}")

# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva t de Student
x = np.linspace(t.ppf(0.001, grados_libertad), t.ppf(0.999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t de Student (GL={grados_libertad})')

# Rellenar la región de rechazo (cola derecha)
x_rechazo = np.linspace(valor_critico_t, t.ppf(0.999, grados_libertad), 100)
y_rechazo = t.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico t
plt.axvline(valor_critico_t, color='red', linestyle='--', label=f'Valor Crítico t = {valor_critico_t:.4f}')

# Marcar el estadístico t calculado
plt.axvline(estadistico_t, color='blue', linestyle='-', label=f'Estadístico t Calculado = {estadistico_t:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución t de Student con Región de Rechazo y Estadístico t (GL={grados_libertad})')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_t > valor_critico_t:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) cae en la región de rechazo (t > {valor_critico_t:.4f}).")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que el tiempo promedio real de entrega a domicilio es significativamente mayor a {mu_0} minutos."
else:
    print(f"El estadístico t calculado ({estadistico_t:.4f}) no cae en la región de rechazo (t > {valor_critico_t:.4f}).")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que el tiempo promedio real de entrega a domicilio es significativamente mayor a {mu_0} minutos."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral de {media_muestral} minutos (o una más extrema, es decir, mayor), si la hipótesis nula (el tiempo promedio real es {mu_0} minutos) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan alta si el tiempo promedio real fuera {mu_0} minutos.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que el tiempo promedio real sea {mu_0} minutos.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La media de llenado es igual a 500 mL (μ = 500)
# H1: La media de llenado es diferente a 500 mL (μ ≠ 500)
# Esta es una prueba de dos colas, ya que se menciona que la máquina podría estar descalibrada,
# lo que implica que el llenado podría ser mayor o menor a 500 mL.
mu_0 = 500  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} mL")
print(f"Hipótesis Alternativa (H1): μ ≠ {mu_0} mL (Prueba de dos colas)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.01
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es conocida, por lo tanto, usamos el estadístico Z
# y la distribución normal estándar.
print("\nEstadístico de prueba: Z")
print("Distribución: Normal estándar")

# Datos de la muestra
n = 60  # Tamaño de la muestra
media_muestral = 498  # mL
desviacion_estandar_poblacional = 5  # mL

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de dos colas con α = 0.01, el valor crítico Z es el valor que deja el α/2 = 0.005
# del área en cada cola de la distribución normal estándar.
valor_critico_z_inferior = norm.ppf(alpha/2)
valor_critico_z_superior = norm.ppf(1 - alpha/2)
print(f"\nValores críticos Z para α = {alpha} (dos colas): {valor_critico_z_inferior:.4f} y {valor_critico_z_superior:.4f}")
print(f"Región de rechazo: Z_calculado < {valor_critico_z_inferior:.4f} o Z_calculado > {valor_critico_z_superior:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico Z: Z = (x̄ - μ0) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)
estadistico_z = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} mL")
print(f"Estadístico Z calculado: {estadistico_z:.4f}")

# Calcular el valor p para una prueba de dos colas
# El valor p es el doble de la probabilidad en la cola correspondiente al estadístico Z calculado.
# Como es de dos colas, calculamos la probabilidad en una cola y la multiplicamos por 2.
# Usamos abs(estadistico_z) para encontrar la probabilidad en la cola más extrema.
valor_p = 2 * norm.cdf(-abs(estadistico_z))
print(f"Valor p (para prueba de dos colas): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
y_rechazo_inferior = norm.pdf(x_rechazo_inferior, 0, 1)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo Inferior (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
y_rechazo_superior = norm.pdf(x_rechazo_superior, 0, 1)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.3, label=f'Región de Rechazo Superior (α/2 = {alpha/2})')


# Marcar los valores críticos Z
plt.axvline(valor_critico_z_inferior, color='red', linestyle='--', label=f'Valor Crítico Z Inferior = {valor_critico_z_inferior:.4f}')
plt.axvline(valor_critico_z_superior, color='red', linestyle='--', label=f'Valor Crítico Z Superior = {valor_critico_z_superior:.4f}')


# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico Z')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_z < valor_critico_z_inferior or estadistico_z > valor_critico_z_superior:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo.")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la media de llenado de las botellas es significativamente diferente a {mu_0} mL."
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo.")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la media de llenado de las botellas es significativamente diferente a {mu_0} mL."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral tan o más extrema que {media_muestral} mL (en cualquier dirección desde {mu_0}), si la hipótesis nula (la media de llenado real es {mu_0} mL) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan diferente de {mu_0} si la media real fuera {mu_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la media real sea {mu_0} mL.")
```

## Prueba de Hipótesis para la Reducción de la Presión Arterial

Un laboratorio afirma que su medicamento reduce la presión arterial en 10 mmHg en promedio. Un estudio con 25 pacientes muestra una reducción promedio de 8 mmHg, con σ = 3 mmHg. Con un nivel de significancia α = 0,05.

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**
2.  **Establecer el nivel de significancia α.**
3.  **Identificar el estadístico de prueba y su distribución.**
4.  **Determinar el valor crítico y la región de rechazo.**
5.  **Calcular el estadístico de prueba y el valor p.**
6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**
7.  **Tomar una decisión estadística y concluir en términos prácticos, explicando el significado del valor p.**

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La reducción promedio es igual a 10 mmHg (μ = 10)
# H1: La reducción promedio es diferente a 10 mmHg (μ ≠ 10) - Usaremos una prueba de dos colas
# si la pregunta no especifica si es mayor o menor. Si sospechamos que es menor, sería de cola izquierda.
# En este caso, como la muestra es 8 (menor que 10) y el cliente podría estar interesado en
# si la reducción es *diferente*, o específicamente *menor*, interpretaremos como prueba de cola izquierda
# basándonos en la muestra. Si la intención fuera solo diferente, sería de dos colas.
# Optaremos por la prueba de cola izquierda dado el contexto de "reducción promedio de 8 mmHg".
# Si la intención fuera probar si la reducción es diferente, cambiaríamos H1 a μ ≠ 10 y usaríamos alpha/2 en cada cola.
mu_0 = 10  # Reducción promedio poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} mmHg")
print(f"Hipótesis Alternativa (H1): μ < {mu_0} mmHg (Prueba de cola izquierda, basada en la media muestral)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es conocida, por lo tanto, usamos el estadístico Z
# y la distribución normal estándar. Aunque n < 30, al conocer sigma, se usa Z.
print("\nEstadístico de prueba: Z")
print("Distribución: Normal estándar")

# Datos de la muestra
n = 25  # Tamaño de la muestra
media_muestral = 8  # mmHg
desviacion_estandar_poblacional = 3  # mmHg

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de cola izquierda con α = 0.05, el valor crítico Z es el valor
# que deja el 5% del área en la cola izquierda de la distribución normal estándar.
valor_critico_z = norm.ppf(alpha)
print(f"\nValor crítico Z para α = {alpha} (cola izquierda): {valor_critico_z:.4f}")
print(f"Región de rechazo: Z_calculado < {valor_critico_z:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico Z: Z = (x̄ - μ0) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)
estadistico_z = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} mmHg")
print(f"Estadístico Z calculado: {estadistico_z:.4f}")

# Calcular el valor p para una prueba de cola izquierda
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo
# o más extremo que el observado, asumiendo que H0 es verdadera.
valor_p = norm.cdf(estadistico_z)
print(f"Valor p (para prueba de cola izquierda): {valor_p:.4f}")

# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico Z
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico Z')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_z < valor_critico_z:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo (Z < {valor_critico_z:.4f}).")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la reducción promedio real de la presión arterial con el medicamento es significativamente menor a {mu_0} mmHg."
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo (Z < {valor_critico_z:.4f}).")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la reducción promedio real de la presión arterial con el medicamento es significativamente menor a {mu_0} mmHg."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral de {media_muestral} mmHg (o una más extrema, es decir, menor), si la hipótesis nula (la reducción promedio real es {mu_0} mmHg) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan baja si la reducción promedio real fuera {mu_0} mmHg.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la reducción promedio real sea {mu_0} mmHg.")
```

## Prueba de Hipótesis para el Puntaje Promedio de un Examen

Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: El promedio de la clase es igual al promedio nacional (μ = 75)
# H1: El promedio de la clase es diferente al promedio nacional (μ ≠ 75)
# Esta es una prueba de dos colas.
mu_0 = 75  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} puntos")
print(f"Hipótesis Alternativa (H1): μ ≠ {mu_0} puntos (Prueba de dos colas)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es conocida, por lo tanto, usamos el estadístico Z
# y la distribución normal estándar.
print("\nEstadístico de prueba: Z")
print("Distribución: Normal estándar")

# Datos de la muestra
n = 36  # Tamaño de la muestra
media_muestral = 78  # puntos
desviacion_estandar_poblacional = 12  # puntos

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de dos colas con α = 0.05, el valor crítico Z es el valor que deja el α/2 = 0.025
# del área en cada cola de la distribución normal estándar.
valor_critico_z_inferior = norm.ppf(alpha/2)
valor_critico_z_superior = norm.ppf(1 - alpha/2)
print(f"\nValores críticos Z para α = {alpha} (dos colas): {valor_critico_z_inferior:.4f} y {valor_critico_z_superior:.4f}")
print(f"Región de rechazo: Z_calculado < {valor_critico_z_inferior:.4f} o Z_calculado > {valor_critico_z_superior:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico Z: Z = (x̄ - μ0) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)
estadistico_z = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} puntos")
print(f"Estadístico Z calculado: {estadistico_z:.4f}")

# Calcular el valor p para una prueba de dos colas
# El valor p es el doble de la probabilidad en la cola correspondiente al estadístico Z calculado.
# Como es de dos colas, calculamos la probabilidad en una cola y la multiplicamos por 2.
# Usamos abs(estadistico_z) para encontrar la probabilidad en la cola más extrema.
valor_p = 2 * norm.cdf(-abs(estadistico_z))
print(f"Valor p (para prueba de dos colas): {valor_p:.4f}")


# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(-4, valor_critico_z_inferior, 100)
y_rechazo_inferior = norm.pdf(x_rechazo_inferior, 0, 1)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo Inferior (α/2 = {alpha/2})')

x_rechazo_superior = np.linspace(valor_critico_z_superior, 4, 100)
y_rechazo_superior = norm.pdf(x_rechazo_superior, 0, 1)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.3, label=f'Región de Rechazo Superior (α/2 = {alpha/2})')


# Marcar los valores críticos Z
plt.axvline(valor_critico_z_inferior, color='red', linestyle='--', label=f'Valor Crítico Z Inferior = {valor_critico_z_inferior:.4f}')
plt.axvline(valor_critico_z_superior, color='red', linestyle='--', label=f'Valor Crítico Z Superior = {valor_critico_z_superior:.4f}')


# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico Z')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_z < valor_critico_z_inferior or estadistico_z > valor_critico_z_superior:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo.")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que el promedio real de la clase es significativamente diferente a {mu_0} puntos."
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo.")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que el promedio real de la clase es significativamente diferente a {mu_0} puntos."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral tan o más extrema que {media_muestral} puntos (en cualquier dirección desde {mu_0}), si la hipótesis nula (el promedio real de la clase es {mu_0} puntos) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan diferente de {mu_0} si el promedio real fuera {mu_0}.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que el promedio real sea {mu_0} puntos.")
```

6.  **Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: La duración promedio de las baterías es igual a 1200 horas (μ = 1200)
# H1: La duración promedio de las baterías es menor a 1200 horas (μ < 1200)
# Esta es una prueba de cola izquierda.
mu_0 = 1200  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} horas")
print(f"Hipótesis Alternativa (H1): μ < {mu_0} horas (Prueba de cola izquierda)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.01
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es conocida, por lo tanto, usamos el estadístico Z
# y la distribución normal estándar.
print("\nEstadístico de prueba: Z")
print("Distribución: Normal estándar")

# Datos de la muestra
n = 50  # Tamaño de la muestra
media_muestral = 1180  # horas
desviacion_estandar_poblacional = 100  # horas

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de cola izquierda con α = 0.01, el valor crítico Z es el valor
# que deja el 1% del área en la cola izquierda de la distribución normal estándar.
valor_critico_z = norm.ppf(alpha)
print(f"\nValor crítico Z para α = {alpha} (cola izquierda): {valor_critico_z:.4f}")
print(f"Región de rechazo: Z_calculado < {valor_critico_z:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico Z: Z = (x̄ - μ0) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)
estadistico_z = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} horas")
print(f"Estadístico Z calculado: {estadistico_z:.4f}")

# Calcular el valor p para una prueba de cola izquierda
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo
# o más extremo que el observado, asumiendo que H0 es verdadera.
valor_p = norm.cdf(estadistico_z)
print(f"Valor p (para prueba de cola izquierda): {valor_p:.4f}")

# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico Z
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico Z')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_z < valor_critico_z:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo (Z < {valor_critico_z:.4f}).")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = f"Hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la duración promedio real de las baterías es significativamente menor a {mu_0} horas."
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo (Z < {valor_critico_z:.4f}).")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = f"No hay evidencia estadística suficiente, con un nivel de significancia del {alpha*100}%, para concluir que la duración promedio real de las baterías es significativamente menor a {mu_0} horas."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print(f"El valor p representa la probabilidad de observar una media muestral de {media_muestral} horas (o una más extrema, es decir, menor), si la hipótesis nula (la duración promedio real es {mu_0} horas) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan baja si la duración promedio real fuera {mu_0} horas.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que la duración promedio real sea {mu_0} horas.")
```

## Prueba de Hipótesis para el Peso Promedio de los Paquetes de Cereal

Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de significancia α = 0,05.

Seguiremos los pasos para realizar la prueba de hipótesis:

1.  **Definir las hipótesis nula y alternativa.**

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# --- Paso 1: Definir las hipótesis nula y alternativa ---
# H0: El peso promedio de los paquetes es igual a 500 g (μ = 500)
# H1: El peso promedio de los paquetes es menor a 500 g (μ < 500)
# Esta es una prueba de cola izquierda.
mu_0 = 500  # Media poblacional bajo la hipótesis nula
print(f"Hipótesis Nula (H0): μ = {mu_0} g")
print(f"Hipótesis Alternativa (H1): μ < {mu_0} g (Prueba de cola izquierda)")

# --- Paso 2: Establecer el nivel de significancia α ---
alpha = 0.05
print(f"\nNivel de significancia (α): {alpha}")

# --- Paso 3: Identificar el estadístico de prueba y su distribución ---
# La desviación estándar poblacional (σ) es conocida, por lo tanto, usamos el estadístico Z
# y la distribución normal estándar.
print("\nEstadístico de prueba: Z")
print("Distribución: Normal estándar")

# Datos de la muestra
n = 40  # Tamaño de la muestra
media_muestral = 495  # g
desviacion_estandar_poblacional = 15  # g

# --- Paso 4: Determinar el valor crítico y la región de rechazo ---
# Para una prueba de cola izquierda con α = 0.05, el valor crítico Z es el valor
# que deja el 5% del área en la cola izquierda de la distribución normal estándar.
valor_critico_z = norm.ppf(alpha)
print(f"\nValor crítico Z para α = {alpha} (cola izquierda): {valor_critico_z:.4f}")
print(f"Región de rechazo: Z_calculado < {valor_critico_z:.4f}")

# --- Paso 5: Calcular el estadístico de prueba y el valor p ---
# Fórmula del estadístico Z: Z = (x̄ - μ0) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(n)
estadistico_z = (media_muestral - mu_0) / error_estandar
print(f"\nError estándar: {error_estandar:.4f} g")
print(f"Estadístico Z calculado: {estadistico_z:.4f}")

# Calcular el valor p para una prueba de cola izquierda
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo
# o más extremo que el observado, asumiendo que H0 es verdadera.
valor_p = norm.cdf(estadistico_z)
print(f"Valor p (para prueba de cola izquierda): {valor_p:.4f}")

# --- Paso 6: Graficar la distribución, región de rechazo, valor crítico y estadístico calculado ---
# Generar puntos para la curva normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # Distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar (Z)')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Marcar el valor crítico Z
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico Z')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- Paso 7: Tomar una decisión estadística y concluir ---
print("\n--- Decisión Estadística y Conclusión ---")
if estadistico_z < valor_critico_z:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo (Z < {valor_critico_z:.4f}).")
    print("Decisión: Rechazar la hipótesis nula (H0).")
    conclusion = "Hay evidencia estadística suficiente, con un nivel de significancia del 5%, para concluir que el peso promedio real de los paquetes de cereal es significativamente menor a 500 g."
else:
    print(f"El estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo (Z < {valor_critico_z:.4f}).")
    print("Decisión: No rechazar la hipótesis nula (H0).")
    conclusion = "No hay evidencia estadística suficiente, con un nivel de significancia del 5%, para concluir que el peso promedio real de los paquetes de cereal es significativamente menor a 500 g."

print(f"\nConclusión: {conclusion}")

# Explicación del valor p
print(f"\nExplicación del valor p ({valor_p:.4f}):")
print("El valor p representa la probabilidad de observar una media muestral de 495 g (o una más extrema, es decir, menor), si la hipótesis nula (el peso promedio real es 500 g) fuera verdadera.")
if valor_p < alpha:
    print(f"Como el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {alpha}), rechazamos la hipótesis nula. Esto sugiere que es poco probable obtener una media muestral tan baja si el peso promedio real fuera 500 g.")
else:
     print(f"Como el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia (α = {alpha}), no rechazamos la hipótesis nula. Esto sugiere que la media muestral observada no es lo suficientemente inusual como para descartar la posibilidad de que el peso promedio real sea 500 g.")
```

**Justificación para usar p = 0.5 sin datos preliminares:**

Cuando no se dispone de información previa sobre la proporción poblacional (p), se utiliza un valor de p = 0.5 para calcular el tamaño de muestra. Esto se debe a que el producto `p * (1-p)` alcanza su valor máximo cuando p = 0.5.

La fórmula del tamaño de muestra para proporciones es:

n = (Z^2 * p * (1-p)) / ME^2

Donde:
- n es el tamaño de muestra.
- Z es el valor crítico de la distribución normal estándar para el nivel de confianza deseado.
- p es la proporción poblacional estimada.
- ME es el margen de error deseado.

El término `p * (1-p)` representa la variabilidad de la proporción. Al usar p = 0.5, este término es 0.5 * (1 - 0.5) = 0.25, que es el valor máximo posible (ningún otro valor de p entre 0 y 1 resultará en un producto mayor).

Al utilizar el valor máximo de `p * (1-p)`, se obtiene el tamaño de muestra más grande posible para un determinado margen de error y nivel de confianza. Esto garantiza que el tamaño de muestra calculado sea suficiente para lograr el margen de error deseado, incluso si la proporción poblacional real es diferente de 0.5. En otras palabras, usar p = 0.5 proporciona una estimación conservadora del tamaño de muestra, asegurando que tengamos suficientes datos para nuestro análisis.
